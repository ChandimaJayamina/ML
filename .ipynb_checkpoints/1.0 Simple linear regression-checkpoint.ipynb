### Import essential libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Load data

```python
df = pd.read_csv('./datafiles/homeprices.csv')
df.head()
```

### Familier with dataset 

```python
%matplotlib inline
plt.scatter(df.area, df.price, color='red', marker='*')
plt.xlabel('Area(sqr.ft)')
plt.ylabel('Price($)')
```

### Creating the model

```python
from sklearn import linear_model
linear_reg = linear_model.LinearRegression()
linear_reg.fit(df[['area']], df['price'])
```

### How regression line look like

```python
%matplotlib inline
plt.xlabel('area', fontsize=20)
plt.ylabel('price', fontsize=20)
plt.scatter(df.area,df.price,color='red',marker='+')
plt.plot(df.area, linear_reg.predict(df[['area']]), color='blue')
```

### Predict using the model

```python
linear_reg.predict([[3200]])
```

### Checking the coefficient and intercept

```python
print('Coefficient: ' + str(linear_reg.coef_[0]) + ', Intercept: ' + str(linear_reg.intercept_))
```
