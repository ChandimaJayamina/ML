```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_digits
digits = load_digits()
df = pd.DataFrame(digits.data)
df['target'] = digits.target
df.head()
```

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits=3)
kf
```

```python
# Lets say we have 10 numbers and lets see how we split it
for train_index, test_index in kf.split([1,2,3,4,5,6,7,8,9]):
    print(train_index, test_index)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.model_selection import cross_val_score
print("Logistic Regression" + str(cross_val_score(LogisticRegression(), digits.data, digits.target)))
```

```python
print("SVC " + str(cross_val_score(SVC(), digits.data, digits.target)))
print("Random Forest " + str(cross_val_score(RandomForestClassifier(), digits.data, digits.target)))
```
