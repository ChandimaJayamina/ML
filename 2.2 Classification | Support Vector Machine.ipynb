####  How to decide which boundry is the best 

#### The nearest data to line and the distance. If the distance is big it is the best line that seperate 2 set. These nearest data points called Support vectors

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
dir(iris)
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target
df.head()
```

```python
df['Flower_name'] = df.target.apply(lambda x : iris.target_names[x])
df.head()
```

### Visualizing data

```python
df0 = df[df.target==0]
df1 = df[df.target==1]
df2 = df[df.target==2]
```

```python
%matplotlib inline
plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'], color='blue')
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'], color='red')
plt.scatter(df2['sepal length (cm)'], df2['sepal width (cm)'], color='green')
```

```python
from sklearn.model_selection import train_test_split
x = df.drop(columns=['target', 'Flower_name'])
y = df.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
from sklearn.svm import SVC
svc_model = SVC()
svc_model.fit(X_train, y_train) # Default C=1, gamma=auto, kernel=rbf
```

```python
svc_model.score(X_test, y_test)
```
