```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./datafiles/salaries.csv')
df.head()
```

```python
df_x = df.drop('salary_more_then_100k', axis='columns')
df_y = df['salary_more_then_100k']
df_x.head()
```

```python
df_y.head()
```

```python
# Need to deal with categorical data
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
```

```python
df_x['company'] = label_encoder.fit_transform(df_x['company'])
df_x['job'] = label_encoder.fit_transform(df_x['job'])
df_x['degree'] = label_encoder.fit_transform(df_x['degree'])
df_x.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df_x, df_y, test_size=0.1)
```

```python
from sklearn import tree
DTree = tree.DecisionTreeClassifier()
```

```python
DTree.fit(X_train, y_train)
```

```python
DTree.score(X_test, y_test)
```
