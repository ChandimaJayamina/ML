#### In decision tree we create one decision tree but in Random forest we get data set and we create multiple datasets and create decision trees for those and we use majority vote

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_digits
digits = load_digits()
dir(digits)
```

```python
# Display images of dataset
%matplotlib inline
plt.gray()
plt.matshow(digits.images[0])
```

```python
df = pd.DataFrame(digits.data)
df['target'] = digits.target
df.head()
```

```python
x = df.drop(columns=['target'])
y = df['target']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
randf_model = RandomForestClassifier() # Default n_estimators(random trees):10 
randf_model.fit(X_train, y_train)
```

```python
randf_model.score(X_test, y_test)
```

```python
y_pred = randf_model.predict(X_test)
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
import seaborn as sns
plt.figure(figsize=(10,7))
sns.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
